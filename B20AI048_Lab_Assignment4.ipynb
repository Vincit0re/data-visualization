## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plot: Monthly changes in different food grain stocks.

```python
# we have to plot the graph for monthly changes in different food grain stocks 
# we are using the data_02 for this purpose
# The same graph needs to be generated for Screen display, Presentation and Poster Print. (Using seaborn set context feature)
# We have to plot all these three side by side using subplots feature
data_02 = pd.read_csv(
    'data/Production_of_Agriculture_Stock-Ministry_of_Finance.csv')
# prepare the data
data_pr = data_02.transpose()
# make first row as column names and drop the first 2 rows
data_pr.columns = data_pr.iloc[0]
data_pr = data_pr.drop(data_pr.index[0:3])

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data_pr.index = data_pr.index.map(lambda x: months[int(x.split('-')[0])-1])

# - Create different color palette
#     - 1 Continuous
#     - 1 one color sequential
#     - 1 Bipolar sequential
#  Use these color palette appropriately with all the above visualizations side by side using subplots

fig, ax = plt.subplots(1, 3, figsize=(30, 8))
sns.set_context('paper')
sns.lineplot(data=data_pr, ax=ax[0], palette='viridis')
sns.set_context('talk')
sns.lineplot(data=data_pr, ax=ax[1], palette='Blues')
sns.set_context('poster')
sns.lineplot(data=data_pr, ax=ax[2], palette='RdBu')
# set figure title
fig.suptitle('Monthly changes in different food grain stocks (from Sep 2013 to Jun 2014)', fontsize=25)
# set subplots title
ax[0].set_title('Paper', fontsize=20)
ax[1].set_title('Talk', fontsize=20)
ax[2].set_title('Poster', fontsize=20)

# set x and y labels
ax[0].set_xlabel('Month', fontsize=15)
ax[0].set_ylabel('Stock', fontsize=15)
ax[1].set_xlabel('Month', fontsize=15)
ax[1].set_ylabel('Stock', fontsize=15)
ax[2].set_xlabel('Month', fontsize=15)
ax[2].set_ylabel('Stock', fontsize=15)
plt.show()
```

# Plot: State wise Colleges of different Ayus Category

```python
# we will use the data_01 for this purpose
# Group the data by State and Ayurveda type
data_01 = pd.read_csv('data/MergerdAyush_Colleges_in_India_01-04-2021_0.csv')
grouped_data = data_01.groupby(['State or Union Territory', 'Type of System']).sum()

# Create a pivot table to reshape the data for plotting
pivot_table = grouped_data.pivot_table(index='State or Union Territory', columns='Type of System', values=['No. of Colleges - Govt', 'No. of Colleges - Non-Govt'])

# - Create different color palette
#     - 1 Continuous
#     - 1 one color sequential
#     - 1 Bipolar sequential
#  Use these color palette appropriately with all the above visualizations side by side using subplots

# plot horizontal bar graph
fig, ax = plt.subplots(1, 3, figsize=(40, 12))
sns.set_context('paper')
sns.set_palette('viridis')
pivot_table.plot(kind='barh', stacked=True, ax=ax[0])
sns.set_context('talk')
sns.set_palette('Blues')
pivot_table.plot(kind='barh', stacked=True, ax=ax[1])
sns.set_context('poster')
sns.set_palette('RdBu')
pivot_table.plot(kind='barh', stacked=True, ax=ax[2])
# set figure title
fig.suptitle('Number of Ayurveda colleges in India', fontsize=25)
# set subplots title
ax[0].set_title('Paper', fontsize=20)
ax[1].set_title('Talk', fontsize=20)
ax[2].set_title('Poster', fontsize=20)
# set x and y labels
ax[0].set_xlabel('Number of colleges', fontsize=15)
ax[1].set_xlabel('Number of colleges', fontsize=15)
ax[2].set_xlabel('Number of colleges', fontsize=15)
plt.show()
```

# Plot: Distribution of National, State and District roads.

```python
data_03 = pd.read_csv('data/Road_Transport_BRS_2017-18_Annexure7_9c.csv')
data_03 = data_03[['Name of the States', 'National Highways', 'State Highways',
       'District Roads']]
# drop the last row
data_03 = data_03.drop(data_03.index[-1])


# Set the index to 'Name of the States'
data_03.set_index('Name of the States', inplace=True)

fig, ax = plt.subplots(1, 3, figsize=(40, 12))
sns.set_context('paper')
sns.set_palette('viridis')
data_03.plot(kind='barh', stacked=True, ax=ax[0])
sns.set_context('talk')
sns.set_palette('Blues')
data_03.plot(kind='barh', stacked=True, ax=ax[1])
sns.set_context('poster')
sns.set_palette('RdBu')
data_03.plot(kind='barh', stacked=True, ax=ax[2])

# set figure title
fig.suptitle('Distribution of National, State and District roads', fontsize=25)
# set subplots title
ax[0].set_title('Paper', fontsize=20)
ax[1].set_title('Talk', fontsize=20)
ax[2].set_title('Poster', fontsize=20)

# set x and y labels
ax[0].set_xlabel('Length of roads (in KM)', fontsize=15)
ax[1].set_xlabel('Length of roads (in KM)', fontsize=15)
ax[2].set_xlabel('Length of roads (in KM)', fontsize=15)

plt.show()
```
